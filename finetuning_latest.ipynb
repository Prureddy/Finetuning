```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the dataset from Hugging Face
dataset = load_dataset("KingNish/reasoning-base-20k", split="train")

# Each sample has: 'user' (prompt), 'assistant' (answer), 'reasoning' (chain-of-thought).
# We'll now combine 'assistant' + 'reasoning' to improve reasoning alignment.

# Load the tokenizer
model_name = "meta-llama/Llama-3.2-1B"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Ensure padding token is set (use EOS token if none)
if tokenizer.pad_token_id is None:
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.pad_token_id = tokenizer.eos_token_id

# Load the model in half-precision (for GPU like T4)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16
)
```

```python
# We're using DeepSpeed for memory optimization, so we keep the model on CPU/GPU as needed.
# No need for explicit device placement here; DeepSpeed will handle it.

# Configure LoRA (Low-Rank Adaptation) for parameter-efficient fine-tuning
lora_config = LoraConfig(
    r=16,             # Bottleneck dimension
    lora_alpha=32,   # Scaling factor
    lora_dropout=0.05, 
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "v_proj"]  # Apply LoRA to query and value projection in attention layers
)
# Wrap the model with LoRA adapters
model = get_peft_model(model, lora_config)
# (Optional) Print trainable parameter count for verification
model.print_trainable_parameters()  # This will show the percentage of trainable params (should be very low).

```

```python
def preprocess_sample(example):
    user_prompt = example["user"].strip()
    assistant = example["assistant"].strip()
    reasoning = example.get("reasoning", "").strip()

    # Merge assistant + reasoning into one target string
    full_answer = f"{assistant}\nReasoning: {reasoning}" if reasoning else assistant

    # Format prompt with clear delimiters
    prompt_text = f"### User:\n{user_prompt}\n\n### Assistant:\n"

    # Tokenize the combined input + answer
    combined = tokenizer(
        prompt_text + full_answer,
        max_length=1024,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    input_ids = combined["input_ids"][0]
    attention_mask = combined["attention_mask"][0]

    # Compute prompt token length to mask label loss
    prompt_tokens = tokenizer(
        prompt_text,
        max_length=1024,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    prompt_len = (prompt_tokens["attention_mask"][0] == 1).sum().item()

    # Mask labels: only compute loss on Assistant's answer
    labels = input_ids.clone()
    labels[:prompt_len] = -100
    labels[attention_mask == 0] = -100

    return {
        "input_ids": input_ids.tolist(),
        "attention_mask": attention_mask.tolist(),
        "labels": labels.tolist()
    }

# Apply preprocessing with clear role markings
processed_dataset = dataset.map(
    preprocess_sample,
    batched=False,
    remove_columns=dataset.column_names
)
```

```python
# Print the first row from the processed dataset
sample = processed_dataset[0]

# Print full keys and sample values (trimmed for readability)
print("Available keys:", sample.keys())
print("input_ids (first 20):", sample["input_ids"][:20])
print("attention_mask (first 20):", sample["attention_mask"][:20])
print("labels (first 20):", sample["labels"][:20])


decoded_input = tokenizer.decode(sample["input_ids"], skip_special_tokens=True)
print("\n Full input to model:\n", decoded_input)

decoded_labels = tokenizer.decode([i for i in sample["labels"] if i != -100], skip_special_tokens=True)
print("\n Expected output (assistant target only):\n", decoded_labels)
```

```python
# # Define training parameters suitable for a 14 GB GPU memory
# batch_size = 2  # per device batch size (small to fit memory)
# gradient_accumulation = 8  # accumulate gradients to simulate a larger batch
# # Effective batch size = batch_size * gradient_accumulation = 16

# training_args = TrainingArguments(
#     output_dir="llama-1b-finetuned-output1",
#     overwrite_output_dir=True,
#     num_train_epochs=3,
#     per_device_train_batch_size=batch_size,
#     gradient_accumulation_steps=gradient_accumulation,
#     fp16=True,  # use 16-bit precision for training
#     bf16=False, # use BF16 if supported; here we assume FP16
#     learning_rate=1e-4,  # LoRA often allows a slightly higher learning rate
#     warmup_steps=100,    # warm up for 100 steps
#     logging_steps=50,
#     save_strategy="no",  # we'll handle saving manually (to include LoRA weights merged)
#     deepspeed="config.json",  # DeepSpeed config file (see config.json below)
#     report_to="none"  # no integrated experiment tracking in this script
# )

# # Initialize the Trainer
# trainer = Trainer(
#     model=model,
#     tokenizer=tokenizer,
#     args=training_args,
#     train_dataset=processed_dataset,
#     eval_dataset=None  # no explicit evaluation dataset provided in this fine-tuning script
# )

# # Start training
# trainer.train()



from transformers import TrainingArguments, Trainer
from datasets import load_dataset

# Optional: split 10% for validation to monitor overfitting
# Split raw dataset first
dataset_split = dataset.train_test_split(test_size=0.1)

# Then apply preprocessing to each split
train_dataset = dataset_split["train"].map(
    preprocess_sample,
    batched=False,
    remove_columns=dataset_split["train"].column_names
)

eval_dataset = dataset_split["test"].map(
    preprocess_sample,
    batched=False,
    remove_columns=dataset_split["test"].column_names
)


# Training config
batch_size = 2
gradient_accumulation = 8  # effective batch size = 16

training_args = TrainingArguments(
    output_dir="llama-1b-finetuned-output1",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=batch_size,
    gradient_accumulation_steps=gradient_accumulation,
    fp16=True,
    bf16=False,
    learning_rate=1e-4,
    warmup_steps=100,
    logging_steps=100,
    lr_scheduler_type="cosine",
    save_strategy="epoch",
    save_total_limit=1,
    deepspeed="config.json",
    report_to="none",
    seed=42,
    remove_unused_columns=False  #  Required to avoid column mismatch errors
)


trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)


# Train the model
trainer.train()
```

```python



# After training, merge LoRA weights into the base model to get a full model checkpoint
print("Merging LoRA weights into base model...")
base_model = trainer.model.merge_and_unload()  # merge LoRA and return the base model
base_model.eval()  # set to inference mode (not strictly necessary)
# Save the full model (including the fine-tuned weights) and tokenizer
output_dir = "llama-1b-finetuned-full-model-3epoch-updatedprompt"
os.makedirs(output_dir, exist_ok=True)
base_model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
print(f"Full model with LoRA weights merged saved to {output_dir}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the saved model and tokenizer
model_path = "llama-1b-finetuned-full-model-3epoch-updatedprompt"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.float16,  # or torch.float32 if you're on CPU
    device_map="auto"           # automatically selects GPU/CPU
)

print(f"Model loaded to: {next(model.parameters()).device}")

model.eval()  # set to inference mode
```

```python
system_prompt = (
    "You are a helpful and knowledgeable assistant. "
    "Answer the user's questions clearly, factually, and concisely. Do not hallucinate. "
    "Avoid making things up, and do not include any additional questions or prompts for the user. "
    "If you're not sure, say so."
    "Do not include reasoning part in the solution."
)

user_prompt = "What is the discriminant of the quadratic equation 5x^2 - 2x + 1 = 0?"

prompt = f"{system_prompt}\n\n### User:\n{user_prompt}\n\n### Assistant:\n"


inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=200,
        do_sample=True,
        temperature=0.1,
        top_p=0.9,
        top_k=50,
        eos_token_id=tokenizer.eos_token_id,
        repetition_penalty=1.1,
        no_repeat_ngram_size=1
    )

response = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Optional: remove the prompt from the response if it appears in output
cleaned_response = response.split("Assistant:")[-1].strip()

print("\n Model Output:\n", cleaned_response)
```
